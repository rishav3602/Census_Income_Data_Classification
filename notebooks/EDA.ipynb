```python
import pandas as pd
```

```python
## Data Ingestions step
data=pd.read_csv('data/census.csv')
data.head()
```

```python
## making a copy of data so that it do not harm my original data file.
df = data.copy()
df.head()
```

```python
## now we will check for the null values in our dataset.
df.isnull().sum()
```

```python
## our data set have no null values and we are good to go
```

```python
df.head()
```

```python
df.columns
```

```python
## we have to change the columns name cause its not appropriate

df.rename(columns = {'39':'age', ' State-gov':'workclass',
                              ' 77516':'fnlwgt',' Bachelors':'education', ' 13':'education-num', ' Never-married':'marital-status', ' Adm-clerical':'occupation', ' Not-in-family':'relationship', ' White':'race', ' Male':'sex', ' 2174':'capital-gain',' 0':'capital-loss', ' 40':'hours-per-week', ' United-States':'country',' <=50K':'income'}, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

```python
## we can drop fnlwgt and education-num cause it will not affect my model accuracy
df=df.drop(labels=['fnlwgt','education-num'],axis=1)
df.head()
```

```python
## we will  check for duplicated records
df.duplicated().sum()
```

```python
## we have 3465 duplicate records so we will drop the duplicated values
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
## now we got rid of duplicated values....!!!
```

```python
## segregate numerical and categorical columns

numerical_columns=df.columns[df.dtypes!='object']
categorical_columns=df.columns[df.dtypes=='object']
print("Numerical columns:",numerical_columns)
print('Categorical Columns:',categorical_columns)
```

```python
df[categorical_columns].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x=0
for i in numerical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x=0
for i in categorical_columns:
    sns.histplot(data=df,x=i,kde=True)
    print('\n')
    plt.show()
```
